```python
#tensor 3D
arrten = np.arange(27).reshape((3,3,3))
arrten
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Define a figure
fig = plt.figure()

# Add 3D subplot
ax = fig.add_subplot(111, projection='3d')

# Creating three arrays x, y, z to create the 3D plot
x = np.repeat(np.arange(arrten.shape[0]), arrten.shape[1]*arrten.shape[2])
y = np.tile(np.repeat(np.arange(arrten.shape[1]), arrten.shape[2]), arrten.shape[0])
z = np.tile(np.arange(arrten.shape[2]), arrten.shape[0]*arrten.shape[1])

# Creating the 3D plot
ax.scatter(x, y, z, c=arrten.flatten(), cmap='cool')  # Here, arrten is the tensor to be visualized

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
#tablica 5D
arr5 = np.arange(240).reshape((2,2,4,5,3))
arr5
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

for a in range(arr5.shape[0]):
    for b in range(arr5.shape[1]):
        # Creating a figure
        fig = plt.figure()

        # Adding 3D subplot
        ax = fig.add_subplot(111, projection='3d')

        # Preparing data for 3D plot
        x = np.repeat(np.arange(arr5[a][b].shape[0]), arr5[a][b].shape[1]*arr5[a][b].shape[2])
        y = np.tile(np.repeat(np.arange(arr5[a][b].shape[1]), arr5[a][b].shape[2]), arr5[a][b].shape[0])
        z = np.tile(np.arange(arr5[a][b].shape[2]), arr5[a][b].shape[0]*arr5[a][b].shape[1])

        # Plotting the data
        ax.scatter(x, y, z, c=arr5[a][b].flatten(), cmap='cool')

        # Setting labels
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')

        plt.show()
```

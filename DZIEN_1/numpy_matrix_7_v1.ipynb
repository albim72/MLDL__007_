```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jv  # jv is the Bessel function

x = np.linspace(0, 10, 100)
for n in range(4):
    plt.plot(x, jv(n, x), label=r'$J_{}(x)$'.format(n))
plt.title("Bessel Functions")
plt.xlabel("x")
plt.legend()
plt.grid(True)
plt.show()
```
